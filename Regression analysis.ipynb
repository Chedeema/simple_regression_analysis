## 2. When and How to Use Regression Models
##### When to Use Regression:
- Prediction: Use regression when you want to predict the value of one variable based on the value(s) of other variable(s).
- Understanding Relationships: Use regression to understand the strength and type (positive or negative) of the relationship between variables.
- Trend Analysis: Use regression to identify trends over time or across various categories.
##### How to Use Regression:
- Simple Linear Regression:
When you have one independent variable and want to predict or understand its effect on the dependent variable.
- Multiple Linear Regression:
When you have multiple independent variables and want to see how they collectively influence the dependent variable.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
# Load the dataset
data = pd.read_csv('marketing_company_data.csv')

# Inspect the first few rows
data.head()
```

```python
# Scatter plot for Budget vs Conversions
sns.scatterplot(x='Budget (USD)', y='Conversions', data=data)
plt.title('Budget vs Conversions')
plt.show()

# Scatter plot for Clicks vs Conversions
sns.scatterplot(x='Clicks', y='Conversions', data=data)
plt.title('Clicks vs Conversions')
plt.show()

# Scatter plot for Impressions vs Conversions
sns.scatterplot(x='Impressions', y='Conversions', data=data)
plt.title('Impressions vs Conversions')
plt.show()
```

```python
# Define the independent (X) and dependent (y) variables
X = data['Budget (USD)']
y = data['Conversions']

# Add a constant to the independent variable (for the intercept)
X = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(y, X).fit()

# Get the summary of the regression model
model.summary()
```

Interpretation of the Output:
- Intercept $\beta_0$: This is the predicted value of Conversions when Budget is 0.
- Slope $\beta_1$: This is the change in Conversions for a one-unit increase in Budget (USD).
- R-squared: This tells you how much variance in Conversions is explained by Budget. An 
$R^2$ close to 1 indicates a strong relationship.
- P-value: If the p-value for Budget (USD) is less than 0.05, it suggests that Budget (USD) is a statistically significant predictor of Conversions.

##### Extract Values from the Table
From the regression results table:

- Intercept $\beta_0$: 231.5751
- Slope $\beta_1$: 0.0033

  
The regression equation is:

$Y=231.5751+0.0033×X$

Where:

Y is the predicted Conversions.
X is the Budget (USD).

```python
# Extract the independent variable (Budget) and dependent variable (Conversions)
X = data['Budget (USD)']
y = data['Conversions']

# Coefficients from the regression model
beta_0 = 231.5751  # Intercept (constant)
beta_1 = 0.0033    # Slope (coefficient for Budget)

# Scatter plot of Budget vs Conversions
sns.scatterplot(x=X, y=y)
plt.title('Budget vs Conversions with Regression Line')

# Generate the trend line using the regression equation: Y = beta_0 + beta_1 * X
y_pred = beta_0 + beta_1 * X

# Plot the trend line
plt.plot(X, y_pred, color='red', label='Trend Line')

# Add a legend
plt.legend()

# Show the plot
plt.xlabel('Budget (USD)')
plt.ylabel('Conversions')
plt.show()
```
