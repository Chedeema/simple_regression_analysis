# Regression Analysis

- Simple Linear Regression: $$Y = \beta_0 + \beta_1X + \epsilon$$
- Multiple Linear Regression: $$Y = \beta_0 + \beta_1X_1 + \beta_2X_2 + \dots + \beta_nX_n + \epsilon$$

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
data = pd.read_csv('marketing_company_data.csv')
data
```

```python
#Define your dependent and independent variables
X = data['Budget (USD)']
Y = data['Conversions']
```

```python
#Add a constant to the independent variable
X = sm.add_constant(X)

#fit the regression model
model = sm.OLS(Y,X).fit()
```

```python
model.summary()
```

- Intercept: $\beta_0$ 
- slope: $\beta_1$

- Intercept is the predicted value of conversion when budget is 0
- Slope is the change in conversions for a non-unit increase in budget
- an r square close to 1 indicates a strong relationship between budget and conversion
- p-value is less than 0.05 it suggest that there is a strong relationship between budget and conversions, therefore a good statistical predictor

$Y = 231.5751 + 0.0033 * X$

```python
#Extract independent and dependent variable
X = data['Budget (USD)']
Y= data['Conversions']

#Coefficients from the regression model
beta_0 = 231.5751 #(Intercept)
beta_1 = 0.0033  #(Slope)

#Generate the trendline using regression euqtion
y_pred = beta_0 + beta_1*X

#Plot the scatter chart
sns.scatterplot(x = X, y = Y)
plt.title('Budget Vs. Conversion')

# Add a trendline
plt.plot(X, y_pred, color = 'red', label = 'Trendline')

plt.legend()

plt.xlabel('Budget (USD)')
plt.ylabel('Conversions')
plt.show()
```
